```python
import fitz
import io
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt

def extract_text_and_images_from_pdf(pdf_path):
    document = pymupdf.open(pdf_path)
    extracted_data = []
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text = page.get_text()
        if len(text) <= 10:
            # Если на странице мало текста, извлекаем изображения и применяем OCR
            pix = page.get_pixmap()
            img = Image.open(io.BytesIO(pix.tobytes(output="png")))
            ocr_text = pytesseract.image_to_string(img, lang='rus')
            extracted_data.append((page_num, ocr_text, img))
    return extracted_data

pdf_path = 'reports/мэйл.ру 2022 год.pdf'
extracted_data = extract_text_and_images_from_pdf(pdf_path)

for page_num, text, img in extracted_data:
    plt.figure()
    plt.imshow(img)
    plt.axis('off')
    plt.title(f'Image from Page {page_num + 1}')
    plt.show()
    print(f"OCR Text from Image on Page {page_num + 1}:\n", text)
```

```python
import pymupdf
import pdfplumber
import tabula
import pandas as pd
import fitz
import io
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt

def extract_text_from_pdf(pdf_path):
    document = pymupdf.open(pdf_path)
    text = ""
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text += page.get_text()
    return text

def extract_text_with_pdfplumber(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ""
        for page in pdf.pages:
            text += page.extract_text()
    return text

def extract_tables_from_pdf(pdf_path):
    tables = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True)
    return tables

def convert_tables_to_dataframe(tables):
    dataframes = []
    for table in tables:
        if isinstance(table, pd.DataFrame):
            dataframes.append(table)
        else:
            # Конвертация списка списков в DataFrame
            df = pd.DataFrame(table[1:], columns=table[0])
            dataframes.append(df)
    return dataframes

def extract_text_and_images_from_pdf(pdf_path):
    document = pymupdf.open(pdf_path)
    extracted_data = []
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text = page.get_text()
        if len(text) <= 10:
            # Если на странице мало текста, извлекаем изображения и применяем OCR
            pix = page.get_pixmap()
            img = Image.open(io.BytesIO(pix.tobytes(output="png")))
            ocr_text = pytesseract.image_to_string(img, lang='rus')
            extracted_data.append((page_num, ocr_text, img))
    return extracted_data

def process_pdf(pdf_path):
    # Извлечение текста из PDF
    pdf_text_mupdf = extract_text_from_pdf(pdf_path)
    pdf_text_plumber = extract_text_with_pdfplumber(pdf_path)
    # Извлечение таблиц из PDF
    pdf_tables = extract_tables_from_pdf(pdf_path)
    # Преобразование таблиц в DataFrame Pandas
    pdf_dataframes = [table for table in pdf_tables if isinstance(table, pd.DataFrame)]
    # Извлечение текста и изображений с использованием OCR
    extracted_data = extract_text_and_images_from_pdf(pdf_path)

    # Объединение всех извлеченных данных в один текст и список DataFrame
    combined_text = pdf_text_mupdf + '\n' + pdf_text_plumber
    combined_tables = pdf_dataframes
    for _, text, _ in extracted_data:
        combined_text += '\n' + text

    return combined_text, combined_tables

# Пример использования функции process_pdf для обработки PDF
pdf_path = 'reports/Управляющая компания Первая 2022 год.pdf'
combined_text, tables = process_pdf(pdf_path)
print(combined_text)  # Выводим первые 500 символов для краткости
```
